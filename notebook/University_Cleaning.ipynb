#### Load Dataset

#### Dataset Inspection

#### Identify Issues in the Dataset
Looking at the dataset in Excel, the following standard changes need to made along with fixing missing and duplicated values:
- Values for the Gender should be "Male" or "Female"
- Values for the InternetAccess column should be "Yes", "No", or "Unknown"

```python
# Check for missing values
print("Missing Values")
print(df.isnull().sum())
print("")

# Check for duplicate rows
print("Duplicate Rows")
print(df.duplicated().sum())
print("")

# View the duplicate records
df[df.duplicated()]
```

#### Data Cleaning

```python
# Fill in missing values
## Fill in missing values for ParentEducation with "Unknown"
df['ParentEducation'] = df['ParentEducation'].fillna('Unknown')

## Fill in missing values for InternetAccess with "Unknown"
df['InternetAccess'] = df['InternetAccess'].fillna('Unknown')

## Fill in missing values for StudyHoursPerWeek with the median
df['StudyHoursPerWeek'] = df['StudyHoursPerWeek'].fillna(df['StudyHoursPerWeek'].median())

## Drop the rows that are missing in FinalGrade
df = df.dropna(subset=['FinalGrade'])

# Standardize columns
## Standardize Gender column
df['Gender'] = df['Gender'].str.strip().str.title()

## Standardize InternetAccess column
df['InternetAccess'] = df['InternetAccess'].replace({'Y': 'Yes', 'N': 'No', 'Unknown': 'Unknown'})
df['InternetAccess'] = df['InternetAccess'].str.strip().str.title()

# Remove duplicates
df = df.drop_duplicates()
```

#### Final Check for Issues

```python
# Check for missing values
print("Missing Values")
print(df.isnull().sum())
print("")

# Check for duplicate rows
print("Duplicate Rows")
print(df.duplicated().sum())
print("")

# View the duplicate records
df[df.duplicated()]
```

#### Final Dataset Inspection

```python
# Dataset dimesnions
print("Dataset Dimensions")
print("Rows and Columns: ", df.shape)
print("")

# Column names
print("Column Names")
print(df.columns)
print("")

# Data types
print("Data Types")
print(df.dtypes)
print("")

# Summary Statistics
print("Summary Statistics")
print(df.describe())
```
